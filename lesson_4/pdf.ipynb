```python
"""4. В группе туристов 8 человек. С помощью жребия они
выбирают шестерых человек, которые должны идти в село в
магазин за продуктами. Какова вероятность того, что турист Д.,
входящий в состав группы, пойдёт в магазин?


Решение
- найти все возможные исходы (по формуле сочетаний C(8,6) = !8/(!6 *!(8-6))
- найти благоприятные исходы (также по формуле сочетаний C(7,5) = !7/(!5 *!(5-7))))
- и чтобы получить вероятность раделить благоприятные исходы на все возможные исходы
"""
import math

total_turist = 8
need_to_go = 6

all_outcomes = math.comb(total_turist, need_to_go)
print(all_outcomes)
good_outcomes = math.comb(total_turist-1, need_to_go-1)
print(good_outcomes)
probability = good_outcomes / all_outcomes
print(probability)
```
