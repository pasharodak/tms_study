Теория графов: направления и основные понятия
Основные понятия
Вершина и ребро Граф G = (V, E) задаёт множество вершин V и множество рёбер E. Рёбра могут быть ориентированными или неориентированными, взвешенными или невзвешенными. Путь — последовательность рёбер, цикл — замкнутый путь.

Степень и связность Степень вершины — число инцидентных ей рёбер. В ориентированном графе выделяют входящую и исходящую степень. Связный граф (для неориентированных) означает, что есть путь между любой парой вершин.

Деревья и леса Дерево — связный ациклический граф. Лес — набор несвязанных деревьев. Каждый дерево на n вершинах имеет ровно n − 1 рёбер и уникальные пути между вершинами.

Матрицы и списки смежности Матрица смежности A размером |V|×|V| хранит веса рёбер, списки смежности для каждой вершины перечисляют её соседей. Списки эффективнее для разреженных графов.

Прочие термины

Планарность: возможность уложить граф без пересечений рёбер

Кликa: множество взаимосвязанных вершин

Покрытие: набор вершин или рёбер, затрагивающих все рёбра или вершины графа

Поток и сечение: понятия из теории максимального потока

Направления теории графов
Алгоритмика на графах изучает методы поиска и оптимизации путей, расчёта потоков, потоковых разрезов, максимальных матчей.

Комбинаторика и эйлерова/гамильтонова теория исследует выборки рёбер (эйлеровы циклы) и обходы всех вершин (гамильтоновы циклы).

Структурная теория изучает планарные графы, раскраску вершин и рёбер, мини- и максимальные подграфы.

Динамика на графах распространяет модели percolation, модели эпидемий и распространения информации.

Прикладные области социальные сети, биоинформатика, базовые сети связи, дорожные и логистические сети, рекомендательные системы.

Основные алгоритмы на графах в реальной жизни
Поиск в ширину (BFS) обходит все вершины слоями от заданной стартовой; находит кратчайшие пути в невзвешенном графе. Применение: поиск связности, уровневые обходы в сетевых протоколах.

Поиск в глубину (DFS) идёт по одному пути до конца, затем откатывается; находит компоненты сильной связности. Применение: проверка на ацикличность, топологическая сортировка.

Dijkstra рассчитывает кратчайшие пути от одной вершины в взвешенном графе без отрицательных весов. Применение: маршрутизация в сетях, навигационные системы.

Bellman–Ford также находит кратчайшие пути, допускает отрицательные веса и обнаруживает отрицательные циклы. Применение: финансовые модели с отрицательными издержками.

Алгоритм A* эвристический поиск кратчайшего пути с оценочной функцией (например, евклидово расстояние). Применение: навигация в робототехнике и играх.

Минимальное остовное дерево – Kruskal: сортирует рёбра по весу, объединяет компоненты в лес с помощью DSU – Prim: растёт дерево от стартовой вершины, добавляя минимальные рёбра Применение: проектирование сетей связи, минимизация затрат на кабели.

Максимальный поток – Edmonds–Karp: находит увеличивающие пути по BFS – Dinic: использует уровневые графы и блокирующие потоки Применение: распределение ресурсов, задачи балансировки нагрузки.

Поиск сильных компонент – Kosaraju: два DFS-прохождения (в прямом и обратном графе) – Tarjan: один DFS с учётом номеров и стека Применение: анализ модулей в программных графах, разделение на независимые подсистемы.

Топологическая сортировка упорядочивает вершины ациклического ориентированного графа. Применение: планирование задач, компиляция зависимостей.

PageRank ранжирует вершины по важности, имитируя случайный блуждающий веб-пользователь. Применение: поисковые системы, аналитика ссылочных структур.

Графовые модели и сети в ML/DL
Встраивание графов (Graph Embedding)
DeepWalk и Node2Vec используют случайные прогулки + Word2Vec для получения эмбеддингов вершин. Применение: рекомендация друзей, кластеризация сообществ.

LINE, HOPE учитывают первую и вторую окрестности, высокую производительность для больших сетей.

Графовые нейронные сети (GNN)
GCN (Graph Convolutional Network) обобщает свёртки на графах, аггрегируя информацию от соседей.

GraphSAGE агрегирует функции соседей по мини-батчам, масштабируем к большим графам.

GAT (Graph Attention Networks) вводит механизм внимания для дифференцированного взвешивания соседей.

GIN (Graph Isomorphism Network) мощная архитектура для задачи графовой классификации с приращениями на уровне вершин.

Успешность применения
Хемоинформатика предсказание свойств молекул на датасете QM9: RMSE улучшается на 5–10% vs классические методы.

Социальные сети детекция спама и ботов с ROC-AUC > 0.95 на реальных графах Twitter.

Рекомендации повышение CTR на 20% для коллаборативных фильтров с GNN-эмбеддингами.

Трафик и логистика прогнозы интенсивности движения с ошибкой MAE < 5% на датасетах городских дорог.

1. Социальная сеть
Модель Пользователи — узлы, дружеские связи — рёбра.

Задача Выявить сообщества в сети с помощью алгоритма Лувена и визуализировать раскраску узлов по сообществам.

```python
import networkx as nx
import matplotlib.pyplot as plt
from community import community_louvain 

# создаём неориентированный граф
SN = nx.Graph()
edges = [
    ('Alice', 'Bob'), ('Alice', 'Carol'), ('Bob', 'Dave'),
    ('Carol', 'Eve'), ('Eve', 'Frank'), ('Dave', 'Frank'),
    ('Gina', 'Helen'), ('Helen', 'Ian'), ('Ian', 'Gina'),
    ('Eve', 'Gina')
]
SN.add_edges_from(edges)

# обнаруживаем сообщества алгоритмом Лувена
partition = community_louvain.best_partition(SN)

# получаем числовую метку сообщества для каждого узла
colors = [partition[node] for node in SN.nodes()]

# располагаем узлы и рисуем
pos = nx.spring_layout(SN, seed=42)
nx.draw(SN, pos,
        with_labels=True,
        node_color=colors,
        cmap=plt.cm.Set3,
        node_size=500,
        edge_color='gray')
plt.title("Социальная сеть: сообщества Louvain")
plt.show()
```

Пояснения

community_louvain.best_partition возвращает словарь {узел: номер_сообщества}.

Цвет узла соответствует его сообществу.

Алгоритм Лувена быстро масштабируется на большие сети и выявляет плотные кластеры.

2. Транспортная сеть
Модель Перекрёстки или остановки — узлы, дороги или маршруты — взвешенные ориентированные рёбра.

Задача Визуализировать сеть и найти кратчайший путь (алгоритм Дейкстры).

```python
import networkx as nx
import matplotlib.pyplot as plt

# создаём ориентированный взвешенный граф
G = nx.DiGraph()
edges = [
    ('A', 'B', 5), ('A', 'C', 10), ('B', 'C', 3),
    ('B', 'D', 2), ('C', 'D', 1), ('D', 'E', 7),
    ('C', 'E', 9)
]
G.add_weighted_edges_from(edges)

# визуализация
pos = nx.spring_layout(G, seed=24)
nx.draw(G, pos,
        with_labels=True,
        node_color='lightgreen',
        edge_color='gray',
        node_size=600)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Транспортная сеть")
plt.show()

# поиск кратчайшего пути от A до E
path = nx.dijkstra_path(G, 'A', 'E')
length = nx.dijkstra_path_length(G, 'A', 'E')
print(f"Кратчайший путь A → E: {path} с общей длиной {length}")
```

Пояснения

nx.dijkstra_path возвращает список узлов, через которые проходит минимальный маршрут.

nx.dijkstra_path_length — суммарный вес пути.

В транспортных приложениях вес ребра может соответствовать времени, расстоянию или стоимости.

3. Граф знаний для рекомендаций
Модель Пользователи, товары и категории — узлы; связи («купил», «относится к») — рёбра.

Задача Построить и визуализировать граф знаний, затем его обходом или эмбеддингом получать рекомендации.

```python
import networkx as nx
import matplotlib.pyplot as plt

# создаём граф знаний
KG = nx.Graph()
KG.add_edges_from([
    ('User1', 'ItemA'), ('User1', 'ItemB'),
    ('User2', 'ItemA'), ('User2', 'CategoryX'),
    ('ItemA', 'CategoryX'), ('ItemB', 'CategoryY'),
    ('CategoryX', 'ProductLine1'), ('CategoryY', 'ProductLine2'),
    ('User3', 'ItemC'), ('ItemC', 'CategoryY')
])

# настраиваем цвета узлов по типу
color_map = []
for node in KG.nodes():
    if node.startswith('User'):
        color_map.append('skyblue')
    elif node.startswith('Item'):
        color_map.append('lightcoral')
    else:
        color_map.append('lightgray')

# визуализация
pos = nx.spring_layout(KG, seed=7)
nx.draw(KG, pos,
        with_labels=True,
        node_color=color_map,
        font_size=8,
        node_size=500,
        edge_color='silver')
plt.title("Граф знаний для рекомендаций")
plt.show()
```

Пояснения

Узлы трёх типов окрашены по-разному: пользователи, товары, категории.

Для рекомендаций популярны подходы Node2Vec или Graph Neural Networks, которые преобразуют узлы в эмбеддинги, учитывающие структуру и семантику связей.

Класс дерево решений для дальнейшего переиспользования в классах RandomForest и GB

```python
import numpy as np

class DecisionTree:
    class Node:
        def __init__(self, *, feature=None, threshold=None, left=None, right=None, value=None, proba=None):
            self.feature = feature
            self.threshold = threshold
            self.left = left
            self.right = right
            self.value = value        # класс-предсказание в листе
            self.proba = proba        # вероятности классов в листе

    def __init__(
        self,
        max_depth=5,
        min_samples_split=2,
        min_impurity_decrease=0.0,
        criterion='gini',
        max_features=None
    ):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_impurity_decrease = min_impurity_decrease
        self.criterion = criterion
        self.max_features = max_features
        self.root = None

    def _impurity(self, y):
        counts = np.bincount(y, minlength=self.n_classes_)
        probs = counts / counts.sum()
        if self.criterion == 'entropy':
            return -np.sum([p * np.log2(p) for p in probs if p > 0])
        # default gini
        return 1 - np.sum(probs ** 2)

    def _best_split(self, X, y):
        best_gain = 0.0
        best_feat, best_thr = None, None
        parent_impurity = self._impurity(y)
        n_samples, n_features = X.shape

        # выбрать подмножество признаков для случайного леса
        if self.max_features is None:
            features = range(n_features)
        elif isinstance(self.max_features, int):
            features = np.random.choice(n_features, self.max_features, replace=False)
        else:  # доля от n_features
            k = max(1, int(self.max_features * n_features))
            features = np.random.choice(n_features, k, replace=False)

        for feat in features:
            thresholds = np.unique(X[:, feat])
            for thr in thresholds:
                mask = X[:, feat] <= thr
                left_size = mask.sum()
                if left_size < self.min_samples_split or left_size == n_samples:
                    continue

                y_left, y_right = y[mask], y[~mask]
                impurity_left = self._impurity(y_left)
                impurity_right = self._impurity(y_right)
                weighted_impurity = (
                    left_size / n_samples * impurity_left
                    + (n_samples - left_size) / n_samples * impurity_right
                )
                gain = parent_impurity - weighted_impurity

                if gain > best_gain and gain >= self.min_impurity_decrease:
                    best_gain, best_feat, best_thr = gain, feat, thr

        return best_feat, best_thr

    def _build_tree(self, X, y, depth=0):
        # создать лист, если условия остановки выполнены
        counts = np.bincount(y, minlength=self.n_classes_)
        predicted = counts.argmax()
        proba = counts / counts.sum()
        node = self.Node(value=predicted, proba=proba)

        if depth < self.max_depth and len(y) >= self.min_samples_split:
            feat, thr = self._best_split(X, y)
            if feat is not None:
                mask = X[:, feat] <= thr
                node.feature, node.threshold = feat, thr
                node.left = self._build_tree(X[mask], y[mask], depth + 1)
                node.right = self._build_tree(X[~mask], y[~mask], depth + 1)
        return node

    def fit(self, X, y):
        X, y = np.array(X), np.array(y)
        self.n_classes_ = len(np.unique(y))
        self.root = self._build_tree(X, y)

    def _traverse(self, x, node):
        if node.feature is None:
            return node
        branch = node.left if x[node.feature] <= node.threshold else node.right
        return self._traverse(x, branch)

    def predict(self, X):
        X = np.array(X)
        return np.array([self._traverse(x, self.root).value for x in X])

    def predict_proba(self, X):
        X = np.array(X)
        return np.vstack([self._traverse(x, self.root).proba for x in X])
```

Применение DecisionTree в RandomForest и GradientBoosting
Ниже показаны два класс-обёртки, которые переиспользуют DecisionTree для построения ансамблей.

1. RandomForestClassifier
Каждый DecisionTree обучается на случайной бутстрэп-выборке

На каждом сплите дерева используется случайный сабсет признаков (max_features)

Предсказание — усреднение вероятностей (или мажоритарное голосование)

```python
import numpy as np
from collections import Counter

class RandomForestClassifier:
    def __init__(
        self,
        n_estimators=10,
        max_depth=5,
        min_samples_split=2,
        min_impurity_decrease=0.0,
        max_features=None,
        bootstrap=True,
        random_state=None
    ):
        self.n_estimators = n_estimators
        self.bootstrap = bootstrap
        self.trees = []
        self.tree_params = dict(
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_impurity_decrease=min_impurity_decrease,
            max_features=max_features
        )
        if random_state is not None:
            np.random.seed(random_state)

    def fit(self, X, y):
        n_samples = X.shape[0]
        self.classes_ = np.unique(y)
        self.n_classes_ = len(self.classes_)
        self.trees = []
        for _ in range(self.n_estimators):
            # бутстрэп-выборка индексов
            if self.bootstrap:
                idxs = np.random.choice(n_samples, n_samples, replace=True)
            else:
                idxs = np.arange(n_samples)
            tree = DecisionTree(**self.tree_params)
            tree.fit(X[idxs], y[idxs])
            self.trees.append(tree)

    def predict_proba(self, X):
        # собираем вероятности от каждого дерева
        probas = np.array([tree.predict_proba(X) for tree in self.trees])
        # усредняем по деревьям
        return probas.mean(axis=0)

    def predict(self, X):
        # вариант: усреднить вероятности и взять argmax
        proba = self.predict_proba(X)
        return np.argmax(proba, axis=1)
```

2. GradientBoostingClassifier (binary)
Бинарная классификация с логарифмической потерей

На каждом шаге строим дерево на псевдо-остатках (градиента)

Итоговое предсказание — сумма по деревьям, переводящаяся в вероятность через сигмоиду

```python
import numpy as np

def _sigmoid(x):
    return 1 / (1 + np.exp(-x))

class GradientBoostingClassifier:
    def __init__(
        self,
        n_estimators=50,
        learning_rate=0.1,
        max_depth=3,
        min_samples_split=2,
        min_impurity_decrease=0.0,
        max_features=None
    ):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.trees = []
        self.tree_params = dict(
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_impurity_decrease=min_impurity_decrease,
            max_features=max_features
        )

    def fit(self, X, y):
        # y ∈ {0,1}
        # инициализируем F0 = log odds
        p = np.clip(np.mean(y), 1e-5, 1-1e-5)
        self.F0 = np.log(p / (1-p))
        F = np.full(y.shape, self.F0)

        self.trees = []
        for _ in range(self.n_estimators):
            # вычисляем псевдо-остатки: градиент логистической функции
            p = _sigmoid(F)
            residuals = y - p

            # обучаем дерево как регрессор на остатков
            tree = DecisionTree(
                criterion='gini',  # критерий не важен, мы используем деревья как регрессор  
                max_depth=self.tree_params['max_depth'],
                min_samples_split=self.tree_params['min_samples_split'],
                min_impurity_decrease=self.tree_params['min_impurity_decrease'],
                max_features=self.tree_params['max_features']
            )
            # Преобразуем остатки в «классы»: 0 или 1 с порогом 0.5
            # Это хак, чтобы использовать классификационные деревья в регрессии
            res_labels = (residuals > 0).astype(int)
            tree.fit(X, res_labels)

            self.trees.append(tree)
            # обновляем F
            # замена: вместо предсказания деревом непрерывного значения берём proba
            update = tree.predict_proba(X)[:,1] - 0.5
            F += self.learning_rate * update

    def predict_proba(self, X):
        F = np.full((X.shape[0],), self.F0)
        for tree in self.trees:
            update = tree.predict_proba(X)[:,1] - 0.5
            F += self.learning_rate * update
        p = _sigmoid(F)
        return np.vstack([1-p, p]).T

    def predict(self, X):
        proba = self.predict_proba(X)
        return (proba[:,1] > 0.5).astype(int)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Загружаем Iris и оставляем только два класса
data = load_iris()
X, y = data.data, data.target
mask = y != 2
X, y = X[mask], y[mask]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=0
)

# 1. RandomForest
rf = RandomForestClassifier(
    n_estimators=20,
    max_depth=4,
    max_features=2,
    random_state=0
)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("RF Accuracy:", accuracy_score(y_test, y_pred_rf))

# 2. GradientBoosting
gb = GradientBoostingClassifier(
    n_estimators=20,
    learning_rate=0.1,
    max_depth=3,
    max_features=2
)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
print("GB Accuracy:", accuracy_score(y_test, y_pred_gb))
```
